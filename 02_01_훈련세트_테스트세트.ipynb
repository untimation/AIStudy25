```python
#      ( 35,  14)  49 
#     2
#   35 ,  14   

from sklearn.neighbors import KNeighborsClassifier      # k-  ()
kn = KNeighborsClassifier() #  

print(fish_data[4]) #  0  5  
```

```python
# 2      (.fit), (.score) .
#    35  14   

train_input = fish_data[:35]
train_target = fish_target[:35] #  0 ~ 34
test_input = fish_data[35:]
test_target = fish_target[35:]  #   35~ 
```
